```python
ngenre = train.genre_id.nunique() #insignificant since this is a reduced training set
mgenre = int(max(train.genre_id.unique()))

dates = train.release_date.unique()
rdates = int(int(max(dates)) - int(min(dates)))
ngenre
```

```python
def sngTitleDist(t1, t2):
    w1 = set(t1.split())
    w2 = set(t2.split())
    k = len(list(w1 & w2))
    return (1 - k/min(len(w1), len(w2)))
```

```python
def albDist(t1, t2):
    w1 = set(t1.split())
    w2 = set(t2.split())
    k = len(list(w1 & w2))
    return (1 - k/min(len(w1), len(w2)))
```

```python
def artDistance(t1, t2):
    w1 = set(t1.split())
    w2 = set(t2.split())
    k = len(list(w1 & w2))
    return (1 - k/min(len(w1), len(w2)))
```

```python
def genreDist(d1, d2, ngenre, mgenre):
    return (d1-d2)/mgenre
```

```python
def dateDist(d1, d2, rdates):
    return abs(d1-d2)/rdates
```

```python
def songDistance(s1, s2):
    dist = 0
    if s1['sng_title'] == s2['sng_title']:
        return dist
    dist += sngTitleDist(s1['sng_title'], s2['sng_title'])
    
    if s1['album_id'] == s2['album_id'] or s1['alb_title'] == s2['alb_title']:
        return dist
    dist += albDist(s1['alb_title'], s2['alb_title'])
    
    if s1['art_name'] == s2['art_name']:
        return dist
    dist += artDist(s1['art_name'], s2['art_name'])
    
    if s1['genre_id'] == s2['genre_id']:
        return dist
    dist += genreDist(s1['genre_id'], s2['genre_id'], ngenre, mgenre)
    
    if s1['release_date'] == s2['release_date']:
        return dist
    dist += dateDist(s1['release_date'], s2['release_date'], rdates)
    
    return dist
    
```

```python
def age_classifier(t):
    t = int(t['user_age'])
    if t <= 22:
        return 0 #'college kid'
    elif t < 30:
        return 1 #'young adult'
    elif t <= 37:
        return 2 #'middle aged'
    elif t < 50:
        return 3 #'p old'
    else:
        return 4 #'oldie goldie'
```

```python
train['user_age'] = train.apply(age_classifier, axis = 1)
test['user_age'] = test.apply(age_classifier, axis = 1)
train.head()
```

```python
def ts_classifier(t):
    t = int(t['ts_listen'])
    if t <= 500:
        return 0 #'early morning'
    elif t < 1000:
        return 1 #'morning'
    elif t <= 1500:
        return 2 #'afternoon'
    elif t < 2000:
        return 3 #'evening'
    else:
        return 4 #'night'
```

```python
train['ts_listen'] = train.apply(ts_classifier, axis = 1)
test['ts_listen'] = test.apply(ts_classifier, axis = 1)
train.head()
```
